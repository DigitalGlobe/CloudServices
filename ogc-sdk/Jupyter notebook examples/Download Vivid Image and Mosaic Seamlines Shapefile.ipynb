# Download Vivid Mosaic Imagery and Mosaic Seamlines Shapefile

In this example we are going to download Vivid imagery over a BBOX and the corresponding mosaic seamlines. To do this, we will be using a Basemaps connect ID which gives us access to Vivid mosaics. **Be sure to update the connect ID in your .ogc-config file to your Basemaps connect ID.**

#### Initialize the Maxar-OGC Module

```python
from Maxar_OGC import Interface

try:
  sw_ogc = Interface() # if .ogc-config was created
except:
  sw_ogc = Interface('https://securewatch.maxar.com/','<connect_id>') # if .ogc-config was not created
```

#### Search and download imagery over a BBOX. The added filter, legacyId LIKE 'Vivid_Standard_%', filters for Vivid Standard imagery.
If you get a list index out of range error, you are probably using SecureWatch connect ID which can't find any Vivid Imagery.

```python
bbox = "39.622615,-105.174866,39.909736,-104.800000"
filter = "(legacyId LIKE 'Vivid_Standard_%')"
search_results = sw_ogc.search(bbox=bbox, filter=filter)

#View the metadata of the images returned. It will be the most recent due to the default stacking profile.
for feature in search_results:
    print(feature['properties'])

#Save the feature ID of an image. We will use this to download the image. 
desired_featureid = search_results[0]['id']

#Save the legacyId of the image to find the corresponding mosaic seamlines.
desired_legacyId = search_results[0]['properties']['legacyId']

#Declare the file path and file name.
output_location = "C:\\Users\\wi008132\\Downloads\\image2.tiff"
    
download_image_by_pixel_count = sw_ogc.download_image_by_pixel_count(bbox=bbox, height=512, width=512, img_format='geotiff', outputpath=output_location, display=True, featurecollection=desired_featureid)
```

#### Search for and download shapefile of mosaic seamlines that correspond to the image and BBOX

```python
#Add typename and download path variables
typename = "MaxarCatalogMosaicSeamlines"
download_path = "C:\\Users\\wi008132\\Downloads\\test.zip"

#Cutting out the last 10 characters grabs the correct product type name from the Legacy ID.
product_id = desired_legacyId[0:-10]
filter = "(product_id = '{}')".format(product_id)

search_results = sw_ogc.search(bbox=bbox, typename=typename, shapefile=False, filter = filter)

#view metadata of seamlines
for i in range(len(search_results)):
    print(search_results[i]['properties'])
    
#Download shapefile of seamlines by adding shapefile=True argument.
sw_ogc.search(bbox=bbox, typename=typename, shapefile=True, filter = filter, download_path=download_path)
```
